문제1
정수를 입력받아 0 이면 "zero" 양수이면 "plus" 음수이면 "minus" 라고 출력하는 프로그램을 작성하시오.

```python
a = input ("정수를 입력하시오")
a = int(a)

if (a == 0) :
  print("zero")
elif (a > 0) :
  print ("plus")
else :
  print("minus")
```

문제2 영문 대문자를 입력받아 "A, B, C, D, E, F" 알맞은 문자를 출력하고 그 외 문자이면 error 라고 출력하는 프로그램을 작성하시오

```python
a = input("영문 대문자를 입력하시오")
a = str(a)

if a == "A" :
  print("Excellent")
elif a == "B" :
  print("Good")
elif a == "C" :
  print("Usually")
elif a == "D" :
  print("Effort")
elif a == "F" :
  print("Failure")
else :
  print("error")
```

문제4. 1~12사이의 정수를 입력받아 평년의 경우 입력받은 월을 입력받아 평년의 경우 해당 월의 날수를 출력하는 프로그램을 작성하시오

```python
xList = [1, 3, 5, 7, 8, 10, 12 ]
yList = [2]
zList = [4, 6, 9, 11]

a = input("1~12 사이의 정수를 입력하시오")
a = int(a)

if a in xList :
  print("30")
elif a in yList :
  print("28")
elif a in zList :
  print("30")
else : 
  print ("오류입니다.")
```

문제5 남자는 'M' 여자는'F'로 나타내기로 하고 18세 이상을 성인이라고 하자 성별과 나이를 입력받아 성인남자, 성인여자, 미성년남자, 미성년여자를 구분하여 출력하는 프로그램을 작성하시오

```python
gender = input("성별을 입력하시오")
age = input("나이를 입력하시오")
age = int(age)
if gender == 'M' and age >= 18 :
  print("MAN")
elif gender == 'M' and age < 18 :
  print("BOY")
elif gender == 'W' and age >= 18 :
  print("WOMAN")
elif gender == 'W' and age < 18 :
  print("GIRL")
else :
  print("오류입니다.")
```

문제6 년도를 입력받아 윤년인지 평년인지 판단하는프로그램을 작성하시오

```python
year = input("년도를 입력하시오")
year = int(year)

if year % 400 == 0 or year % 4 == 0 and year % 100 != 0 :
  print ("leap year")
else :
  print ("common year")
```
