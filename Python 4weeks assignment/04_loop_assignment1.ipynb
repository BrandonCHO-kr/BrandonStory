100 이하의 양의 정수만 입력된다. while 문을 이용하여 1부터 입력받은 정수까지의 합을 구하여 출력하는 프로그램을 작성하라

```python
n = int(input("숫자를 입력하시오"))
sum = 0
i = 1

while i <= n : ## 1에서 부터 내가 입력한 n값까지 돌아라 (참이면)
    sum += i
    i += 1
print(sum)
```

100 이하의 정수를 입력받아서 입력받은 정수부터 100까지의 합을 출력하는 프로그램을 작성하시오


```python
n = int(input("숫자를 입력하시오"))
sum = 0
## while이 조건이 참이면 계속 반복해라
## 입력한 값에서 부터 100까지 반복해라
while n <= 100 : 
  sum = sum + n
  n = n + 1 ## 증감식 (for 의 증감식)  99 
print(sum)

    
```

한 개의 자연수를 입력받아 그 수의 배수를 차례로 10개 출력하는 프로그램을 작성하시오



```python
n = int(input("자연수를 입력하시오"))

if n < 0 :
  print("잘못된 입력입니다.")
else :
  for i in range(1 , 11) : # 1 ~ 10 까지 i 값에 넣어줌
    print(n * i)


```

100 이하의 양의 정수만 입력된다.
while 문을 이용하여 1부터 입력받은 정수까지의 합을 구하여 출력하는 프로그램을 작성하시오   --- * 문제 1번과 동일


```python
n = int(input("숫자를 입력하시오"))
sum = 0
i = 1

while i <= n : ## 1에서 부터 내가 입력한 n값까지 돌아라 (참이면)
    sum += i
    i += 1
print(sum)
```

0이 입력될 때까지 정수를 계속 입력받아/ 3의 배수와 5의 배수를 제외한 수들의 개수를 출력하는 프로그램을 작성하시오

```python
sum = 0


while True:
  num = int(input("숫자를 입력하시오"))
  if num == 0:
    break
  else:
    if num%3 != 0 and num%5 != 0:
        sum = sum +1

print(sum)
```

삼각형의 밑변의 길이와 높이를 입력받아 넓이를 출력하고, "Continue?"에서 하나의 문자를 입력받아 그 문자가 'Y' 나 'y' 이면 작업을 반복하고 다른 문자이면 종료하는 프로그램을 작성하시오 (넓이는 반올림하여 소수 첫째자리까지 출력한다.)

```python
while True :

  a = int(input("Base = "))
  b = int(input("Height = "))
  result = (a * b) / 2
  print ("Triangle width = " , result)
  c = str(input("Continue ? "))
  
  if c != 'y' and c != 'Y' :
    break
```
