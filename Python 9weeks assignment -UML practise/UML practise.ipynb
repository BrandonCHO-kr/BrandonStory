UML 문제1

```python
class Account :
  def __init__(self, account = "441-0290-1203" , balance = 500000, interestRate = 0.073) :
    self.__account = account
    self.__balance = balance
    self.__interestRate = interestRate
  
  def getAccount(self) :
    return self.__account 
  def setAccount(self, account) :
    self.__account = account
  
  def getBalance (self) :
    return self.__balance
  def setBalance (self, balance) :
    self.__balance = balance
    
  def deposit(self, amount) :
    self.__balance += amount
    return self.__balance

  def calculateInterest (self) :
    a = self.__balance * self.__interestRate
    print("이자율", a)


obj = Account()
print("계좌정보", obj.getAccount() , "현재잔액", obj.getBalance())
obj.deposit(20000)
print("계좌정보", obj.getAccount() , "현재잔액", obj.getBalance())

obj.calculateInterest()
```

```python
# "SQL Plus," 50000, 5
# "Java 2.0", 40000, 3
# "JSP Servlet", 60000, 6

class book :
  def __init__(self, bookName = "defalt",bookPrice = 0 ,bookDiscountRate = 0  ) :
    self.__bookName = bookName
    self.__bookPrice = bookPrice
    self.__bookDiscountRate = bookDiscountRate

  def getBookName (self) :
    return self.__bookName
  def setBookName (self, bookName) :
    self.__bookName = bookName
  
  def getBookPrice (self) :
    return self.__bookPrice
  def setBookPrice (self, bookPrice) :
    self.__bookPrice = bookPrice

  def getBookDiscountRate (self) :
    return self.__bookDiscountRate
  def setBookDiscountRate (self, bookDiscountRate) :
    self.__bookDiscountRate = bookDiscountRate
    
  def BookDiscountPrice(self):
    return self.__bookPrice * ((100 - self.__bookDiscountRate) / 100)

obj1 = book()
obj2 = book()
obj3 = book()

obj1.setBookName("SQL Plus")
obj1.setBookPrice(50000)
obj1.setBookDiscountRate(5)

obj2.setBookName("Java 2.0")
obj2.setBookPrice(40000)
obj2.setBookDiscountRate(3)

obj3.setBookName("JSP Servlet")
obj3.setBookPrice(60000)
obj3.setBookDiscountRate(6)


for i in range(3):
    if i == 0:
        print(f"{obj1.getBookName()} {obj1.getBookPrice()}원 {obj1.getBookDiscountRate()}%")
    elif i == 1:
        print(f"{obj2.getBookName()} {obj2.getBookPrice()}원 {obj2.getBookDiscountRate()}%")
    else:
        print(f"{obj3.getBookName()} {obj3.getBookPrice()}원 {obj3.getBookDiscountRate()}%")


sum = obj1.getBookPrice() + obj2.getBookPrice() + obj3.getBookPrice()
discount = obj1.BookDiscountPrice() + obj2.BookDiscountPrice() + obj3.BookDiscountPrice()

print()
print(f"책 가격의 합: {sum}원")
print(f"할인 된 책 가격의 합: {discount}원")
```
