문제1
다음 출력 예와 같이 출력되는 프로그램을 작성하시오.
합계와 평균은 수식을 이용하세요

```python
kor = 90
mat = 80
eng = 100
kor=int(kor)
mat=int(mat)
eng=int(eng)
sum = kor+mat+eng
avg = (kor+mat+eng)/3
print("kor",kor)
print("mat",mat)
print("eng",eng)
print("sum",sum)
print("avg",int(avg))
```

문제4
세 개의 정수를 입력받아 합과 평균을 출력하는 프로그램을 작성하시오 (단 평균은 소수 이하를 버림하여 정수 부분만 출력한다.)

```python
x = input("첫 번째 정수 20을 입력하시오")
y = input("두 번째 정수 50을 입력하시오")
z = input("세 번째 정수 100을 입력하시오")
x = int(x)
y = int(y)
z = int(z)
sum = x + y + z
avg = (x + y + z)/3

print("sum = ",sum)
print("avg = ",int(avg))
```

문제5
실수의 yard(야드)를 입력받아 cm(센티미터)로 환산하여 입력값과 환산한 값을 출력 예와 같이 소수 둘째자리에서 반올림하여 첫째자리까지 출력하는 프로그램을 작성하시오 (단 1야드 = 91.44cm로 입력은 "yard?"라고 먼저 출력하고, 실수를 입력받는다. 실수는 "double"로 한다.


```python
double = input("yard?")
double = float(double)

yard_1 = 91.44

yard = yard_1 * double

print("10.1yard =",round(yard,1),"cm" )

```
